## Heatmap of total crashes each month 2013-2022

```python
# Filter date range
start = pd.Timestamp('2013-01-01')
end = pd.Timestamp('2022-12-31')
df_filtered = df_crash[(df_crash['CRASH DATE'] >= start) & (df_crash['CRASH DATE'] <= end)]

# Add year and month columns
df_filtered['year'] = df_filtered['CRASH DATE'].dt.year
df_filtered['month'] = df_filtered['CRASH DATE'].dt.month

# Count crashes per month
df_total = df_filtered.groupby(df_filtered['CRASH DATE'].dt.to_period('M')).size().reset_index(name='n_crashes')

# Rename and convert to timestamp
df_total.columns = ['month', 'n_crashes']
df_total['month'] = df_total['month'].dt.to_timestamp()

# Count crashes per year-month
df_total = df_filtered.groupby(["year", "month"]).size().reset_index(name="n_crashes")

# Unstack into 2D matrix
df_cross = df_total.set_index(["year", "month"])["n_crashes"].unstack()

MONTH_LABELS = ["Jan","Feb","Mar","Apr","May","Jun",
                "Jul","Aug","Sep","Oct","Nov","Dec"]

fig = px.imshow(
    df_cross,
    height=350,
    aspect="auto",
    color_continuous_scale="inferno",   # high-contrast: white → yellow → red
    labels={"x": "Month", "y": "Year", "color": "Crashes"},
    title="NYC Crash Frequency by Month and Year (2013–2022)",
    x=MONTH_LABELS,
)

fig.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white",
    font=dict(family="Times New Roman", color="#222222", size=12),
    title_font=dict(size=16, color="#222222"),
    margin=dict(l=60, r=60, t=60, b=50),
    coloraxis_colorbar=dict(
        title="Crashes",
        tickfont=dict(size=11),
        title_font=dict(size=12),
    ),
    xaxis=dict(title="Month", tickfont=dict(size=12)),
    yaxis=dict(title="Year", tickfont=dict(size=12), dtick=1),
)
fig.update_traces(
    textfont=dict(size=9, color="black"),
    hovertemplate="<b>%{y}  %{x}</b><br>Crashes: %{z:,}<extra></extra>",
)
fig.show()

fig.write_html("heatmap_crashes.html")
```

## Heatmap of flight passengers each month 2013-2022

```python
df_tourism = pd.read_csv(r"C:\Users\emmaa\OneDrive - Danmarks Tekniske Universitet\Y4\Social Data and Vizualisation\PANYNJ_Airport_Traffic_Data.csv", low_memory=False)
```

```python
#filter for only arriving passengers and only JFK and LaGardia
df_tourism = df_tourism[df_tourism["Direction"] == "Inbound"]
df_tourism = df_tourism[df_tourism["Airport Code"].isin(["JFK", "LGA"])]

# Define date range
start = "2013-01-01"
end = "2022-12-31"

# Filter dates
df_tourism = df_tourism[
    (df_tourism["Activity Period"] >= start) &
    (df_tourism["Activity Period"] <= end)
]

#make a new table with only month and total passengers per month
monthly_visitors = (
    df_tourism.groupby("Activity Period")["Revenue Passenger Volume"]
    .sum()
    .reset_index()
)
monthly_visitors["Activity Period"] = pd.to_datetime(
    monthly_visitors["Activity Period"]
)

# Add year and month columns
df_tourism = df_tourism.copy()
df_tourism['year'] = pd.to_datetime(df_tourism['Activity Period']).dt.year
df_tourism['month'] = pd.to_datetime(df_tourism['Activity Period']).dt.month

# Aggregate by year and month
df_agg = df_tourism.groupby(['year', 'month'])['Revenue Passenger Volume'].sum().reset_index()

#plot the heatmap
df_cross = df_agg.set_index(['year', 'month'])['Revenue Passenger Volume'].unstack()

MONTH_LABELS = ["Jan","Feb","Mar","Apr","May","Jun",
                "Jul","Aug","Sep","Oct","Nov","Dec"]

fig = px.imshow(
    df_cross,
    height=350,
    aspect="auto",
    color_continuous_scale="inferno",   # high-contrast: white → yellow → red
    labels={"x": "Month", "y": "Year", "color": "Passengers"},
    title="Number of passengers on flights to New York City by Month and Year (2013–2022)",
    x=MONTH_LABELS,
)

fig.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white",
    font=dict(family="Times New Roman", color="#222222", size=12),
    title_font=dict(size=16, color="#222222"),
    margin=dict(l=60, r=60, t=60, b=50),
    coloraxis_colorbar=dict(
        title="Passengers",
        tickfont=dict(size=11),
        title_font=dict(size=12),
    ),
    xaxis=dict(title="Month", tickfont=dict(size=12)),
    yaxis=dict(title="Year", tickfont=dict(size=12), dtick=1),
)
fig.update_traces(
    textfont=dict(size=9, color="black"),
    hovertemplate="<b>%{y}  %{x}</b><br>Passengers: %{z:,}<extra></extra>",
)
fig.show()
fig.write_html("heatmap_tourism.html")
```

## Scatter map of all vehicle collisions in New York

```python
# Parse LOCATION string into separate lat and lon columns, there is a column for latitude and longitude, but it is missing a lot of values and location is more consistently used
df_crash[['lat', 'lon']] = (df_crash['LOCATION']
                             .str.replace('(', '', regex=False)
                             .str.replace(')', '', regex=False)
                             .str.split(',', expand=True)
                             .astype(float))

#map attributes
MAP_STYLE  = 'carto-darkmatter'
NYC_CENTER = dict(lat=40.73, lon=-73.9207)
MAP_ZOOM   = 11

# Sample for performance 200 000, as the point is to show that crashes happen pretty much all over, and this pattern is realized with this size sample, the data is sampled
df_map = df_crash.dropna(subset=['lat', 'lon']).sample(200000, random_state=42)

#create visualization
fig = px.scatter_mapbox(
    df_map,
    lat='lat', lon='lon',
    zoom=MAP_ZOOM,
    height=650,
    center=NYC_CENTER,
)

#add legend and title
fig.update_layout(
    mapbox_style=MAP_STYLE,
    mapbox=dict(center=NYC_CENTER, zoom=MAP_ZOOM, bearing=0, pitch=0),
    title=dict(
        text='<b>NYC Vehicle Collision Locations</b><br>'
             '<sup>Each dot = one recorded collision (2013–2022, random sample of 200 000)</sup>',
        x=0.01, xanchor='left',
        font=dict(size=17, color='#F6C600'),
    ),
    margin=dict(l=0, r=0, t=70, b=0),
)

fig.update_traces(
    marker=dict(size=3, color='#F6C600', opacity=0.5),
    hovertemplate='<b>(%{lat:.4f}, %{lon:.4f})</b><extra></extra>',
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the date interval here:
start_date = '2016-01-01'
end_date = '2020-12-29'

# Ensure 'CRASH DATE' is in datetime format
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], errors='coerce')

# Filter data based on the interval
mask = (df['CRASH DATE'] >= start_date) & (df['CRASH DATE'] <= end_date)
df_filtered = df.loc[mask]

# Group by date and count accidents
daily_crashes = df_filtered.groupby('CRASH DATE').size()

# Calculate the monthly average using resample
# 'ME' stands for Month End
monthly_avg_crashes = daily_crashes.resample('ME').mean()

# Create the plot
plt.figure(figsize=(15, 7))

# Plot the average
monthly_avg_crashes.plot(kind='line', color='#D32F2F', marker='o', linestyle='-')

plt.title('Average number of car accidents per month (2012 to present)')
plt.xlabel('Year')
plt.ylabel('Avg. number of accidents per day in the month')

# Dashed grid
plt.grid(True, linestyle='--', alpha=0.7)

# plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Filter data for the years 2016 to 2022 and months April (4) to October (10)
df_crashes['CRASH DATE'] = pd.to_datetime(df_crashes['CRASH DATE'], errors='coerce')
mask = (df_crashes['CRASH DATE'].dt.year >= 2016) & \
       (df_crashes['CRASH DATE'].dt.year <= 2022) & \
       (df_crashes['CRASH DATE'].dt.month >= 4) & \
       (df_crashes['CRASH DATE'].dt.month <= 10)

df_filtered_period = df_crashes.loc[mask].copy()

# 1. Calculate total number of accidents per borough
total_crashes_by_borough = df_filtered_period.groupby('BOROUGH').size()

# 2. Calculate sum of persons killed by type per borough
category_killed = df_filtered_period.groupby('BOROUGH')[['NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']].sum()

# 3. Calculate ratio (number of deaths / total accidents)
killed_ratio = category_killed.div(total_crashes_by_borough, axis=0)

# 4. Calculate percentage distribution within each borough for labels
killed_pct = category_killed.div(category_killed.sum(axis=1), axis=0) * 100

# Prepare data for plotting
killed_ratio_plot = killed_ratio.rename(columns={
    'NUMBER OF PEDESTRIANS KILLED': 'Pedestrians',
    'NUMBER OF CYCLIST KILLED': 'Cyclists',
    'NUMBER OF MOTORIST KILLED': 'Motorists'
}).reset_index()

killed_pct_plot = killed_pct.rename(columns={
    'NUMBER OF PEDESTRIANS KILLED': 'Pedestrians',
    'NUMBER OF CYCLIST KILLED': 'Cyclists',
    'NUMBER OF MOTORIST KILLED': 'Motorists'
}).reset_index()

# Define colors
colors = {'Motorists': '#B8B4BD', 'Cyclists': '#363B43', 'Pedestrians': '#F9A03F'}

# Create interactive stacked bar chart with Plotly
fig = go.Figure()

for category in ['Motorists', 'Cyclists', 'Pedestrians']:
    fig.add_trace(go.Bar(
        name=category,
        x=killed_ratio_plot['BOROUGH'],
        y=killed_ratio_plot[category],
        marker_color=colors[category],
        # Add text labels for percentages inside bars
        text=killed_pct_plot[category].apply(lambda x: f'{x:.1f}%' if x > 0 else ''),
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(color='white', size=10),
        # Enable custom hover information
        customdata=killed_pct_plot[category],
        hovertemplate="<b>%{x}</b><br>" +
                      f"{category}: " + "%{customdata:.1f}%<extra></extra>"
    ))

# Update layout for stacking and titles
fig.update_layout(
    barmode='stack',
    title='Ratio of Fatalities per Accident (2016-2022, Apr-Oct)',
    xaxis_title='Borough',
    yaxis_title='Ratio (Deaths / Total Accidents)',
    legend_title='Victim Type',
    template='plotly_white',
    # Enable hovermode for interaction
    hovermode='closest'
)

# Save and show
fig.write_html("ratio_total_casualties_analysis_summer.html")
#fig.show()
```

# Investigation correlation between contributing factors with linear regression

```python
from itertools import combinations

#CONSIDER DISPLAYING THIS AS A HEAT MAP FOR READABILITY

# Additions to define cleaned_factor_names and master_factors_df
# 1. Define cleaned_factor_names from the top contributing factors identified previously.

cleaned_factor_base_names = df_crash_winter_top10.head(5)['CONTRIBUTING FACTOR VEHICLE 1'].tolist()

# 2. Prepare master_factors_df with daily crash counts for each of these factors.
#    Use df_crash_winter_filtered for consistency with previous analysis.
df_factors_daily_counts = df_crash_winter_filtered.copy()

# Create binary columns for each contributing factor, suffixed for daily counts
# The column name in master_factors_df will be 'FactorName_DAILY_CRASH_COUNT'
for factor_name in cleaned_factor_base_names:
    col_name = f"{factor_name}_DAILY_CRASH_COUNT"
    df_factors_daily_counts[col_name] = (df_factors_daily_counts['CONTRIBUTING FACTOR VEHICLE 1'] == factor_name).astype(int)

# Group by date and sum the daily crash counts for each factor
master_factors_df = df_factors_daily_counts.groupby('date')[
    [f"{factor_name}_DAILY_CRASH_COUNT" for factor_name in cleaned_factor_base_names]
].sum().reset_index()

# Update cleaned_factor_names to reflect the actual column names in master_factors_df
cleaned_factor_names = [f"{factor_name}_DAILY_CRASH_COUNT" for factor_name in cleaned_factor_base_names]

# 3. Define the lin_reg function as it's used later in the cell.

def lin_reg(x, y):

    # Perform linear regression: y = ax + b
    a, b = np.polyfit(x, y, 1)
    return a, b


# Get all unique combinations of the cleaned factor column names
factor_combinations_columns = list(combinations(cleaned_factor_names, 2))

num_plots = len(factor_combinations_columns)
cols = 3 # Number of columns for the plot grid
rows = (num_plots + cols - 1) // cols # Calculate rows needed

fig, axes = plt.subplots(rows, cols, figsize=(cols * 8, rows * 7))
axes = axes.flatten() # Flatten the axes array for easy iteration

r2_values_factor_correlations = {}

# Define the custom color palette
scatter_color = '#343843'
line_color = '#fda736'

for i, (factor1_col, factor2_col) in enumerate(factor_combinations_columns):
    x = master_factors_df[factor1_col]
    y = master_factors_df[factor2_col]

    a, b = lin_reg(x, y)

    # Calculating R2
    y_hat = a * x + b
    ss_res = np.sum((y - y_hat) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    R2 = 1 - (ss_res / ss_tot) if ss_tot > 0 else 0
    r2_values_factor_correlations[f'{factor1_col} vs {factor2_col}'] = R2

    axes[i].scatter(x, y, s=10, alpha=0.6, color=scatter_color)
    axes[i].axline((0, b), slope=a, color=line_color, label=f'y = {a:.2f}x + {b:.2f}')
    axes[i].legend(loc='upper left', fontsize=8)
    axes[i].set_title(f'{factor1_col.replace("_DAILY_CRASH_COUNT", "").replace("_", " ")} vs\n{factor2_col.replace("_DAILY_CRASH_COUNT", "").replace("_", " ")} (R2 = {R2:.4f})', fontsize=16)
    axes[i].set_xlabel(factor1_col.replace("_DAILY_CRASH_COUNT", "").replace("_", " "), fontsize=10)
    axes[i].set_ylabel(factor2_col.replace("_DAILY_CRASH_COUNT", "").replace("_", " "), fontsize=10)
    axes[i].tick_params(axis='both', labelsize=7)
    axes[i].grid(True, linestyle=':', alpha=0.7)

# Hide any unused subplots
for j in range(num_plots, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Correlation Between Top Contributing Factors (Daily Crash Counts)', fontsize=24, y=1.02)
plt.tight_layout()
# plt.show()

print("\nR-squared values for contributing factor correlations:")
for combo, r2 in r2_values_factor_correlations.items():
    print(f"{combo}: R2 = {r2:.4f}")

plt.savefig("correlation_contributingfactors.png", dpi=150, bbox_inches='tight')
```

```python
# ── Rename columns ────────────────────────────────────────────────────
df_crash = df_crash.rename(columns={
    'LATITUDE':                      'lat_raw',
    'LONGITUDE':                     'lon_raw',
})

# ── Parse dates ───────────────────────────────────────────────────────
df_crash['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y', errors='coerce')
df_crash['year']       = df_crash['CRASH DATE'].dt.year
df_crash['month']      = df_crash['CRASH DATE'].dt.month
df_crash['hour']       = pd.to_datetime(df['CRASH TIME'], format='%H:%M', errors='coerce').dt.hour

# ── Filter to 2014–2022 ───────────────────────────────────────────────
df_crash = df_crash[(df['year'] >= 2014) & (df_crash['year'] <= 2022)].copy()

# ── Parse coordinates ─────────────────────────────────────────────────
loc_parsed      = df_crash['LOCATION'].str.extract(r'\(\s*([\d.\-]+)\s*,\s*([\d.\-]+)\s*\)')
df_crash['latitude']  = pd.to_numeric(loc_parsed[0], errors='coerce')
df_crash['longitude'] = pd.to_numeric(loc_parsed[1], errors='coerce')

lat_fallback    = pd.to_numeric(df_crash['lat_raw'], errors='coerce').replace(0, np.nan)
lon_fallback    = pd.to_numeric(df_crash['lon_raw'], errors='coerce').replace(0, np.nan)
df_crash['latitude']  = df_crash['latitude'].fillna(lat_fallback)
df_crash['longitude'] = df_crash['longitude'].fillna(lon_fallback)

# ── Numeric columns ───────────────────────────────────────────────────
for c in ['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED']:
    df_crash[c] = pd.to_numeric(df[c], errors='coerce').fillna(0).astype(int)

# ── Severity label ────────────────────────────────────────────────────
df_crash['severity'] = np.where(df['NUMBER OF PERSONS KILLED']  > 0, 'Fatal',
                 np.where(df['NUMBER OF PERSONS INJURED'] > 0, 'Injury', 'Property Only'))

# ── Borough & cause ───────────────────────────────────────────────────
df_crash['BOROUGH']   = df_crash['BOROUGH'].fillna('Unknown').str.title()
df_crash['cause'] = df_crash['CONTRIBUTING FACTOR VEHICLE 1'].astype(str).str.strip()
df_crash['cause'] = df_crash['cause'].replace({'nan': np.nan, 'Unspecified': np.nan})

top_causes  = df_crash['cause'].value_counts().nlargest(12).index.tolist()
df_crash = df_crash[df_crash['cause'].isin(top_causes)]

# ── month as string so Plotly treats it as a category, not a number ────
df_crash['month_str'] = df_crash['month'].astype(str).str.zfill(2)


# ── Sortable + readable month label: "01 Jan" sorts Jan→Dec, reads clearly ──
MONTH_NAMES = {
    1:'01 Jan', 2:'02 Feb', 3:'03 Mar', 4:'04 Apr',
    5:'05 May', 6:'06 Jun', 7:'07 Jul', 8:'08 Aug',
    9:'09 Sep',10:'10 Oct',11:'11 Nov',12:'12 Dec',
}
df_crash['month_label'] = df_crash['month'].map(MONTH_NAMES)

# ── Season column for reference ────────────────────────────────────────
WINTER_MONTHS = {11, 12, 1, 2, 3}
df_crash['season'] = df_crash['month'].apply(lambda m: 'Winter (Nov–Mar)' if m in WINTER_MONTHS else 'Summer (Apr–Oct)')
```

```python
# ── Config ────────────────────────────────────────────────────────────
MAP_STYLE       = 'carto-darkmatter'
NYC_CENTER      = dict(lat=40.71, lon=-73.99)
MAP_ZOOM        = 13
SAMPLE_PER_MONTH = 50000


CAUSE_COLORS = {
    'Driver Inattention/Distraction':  '#FF6B6B',
    'Failure to Yield Right-of-Way':   '#FFD93D',
    'Following Too Closely':           '#6BCB77',
    'Unsafe Speed':                    '#4D96FF',
    'Backing Unsafely':                '#C77DFF',
    'Passing or Lane Usage Improper':  '#FF9A3C',
    'Turning Improperly':              '#00C9A7',
    'Traffic Control Disregarded':     '#F72585',
    'Passing Too Closely':             '#7209B7',
    'Aggressive Driving/Road Rage':    '#FF4D6D',
    'Reaction to Uninvolved Vehicle':  '#4CC9F0',
}
GLOBAL_FONT = 'Times New Roman'

# ── Sample per month ───────────────────────────────────────────────────
df_map = df_crash[df_crash['latitude'].notna() & df_crash['longitude'].notna()].copy()

df_map = (
    df_map
    .groupby('month_label', group_keys=False)
    .apply(lambda x: x.sample(min(len(x), SAMPLE_PER_MONTH), random_state=42))
)

print(f'Total points: {len(df_map):,}')

#layout
def apply_layout(fig):
    fig.update_layout(
        mapbox=dict(
            style=MAP_STYLE,
            center=NYC_CENTER,
            zoom=MAP_ZOOM,
        ),
        height=750,
        margin=dict(l=0, r=0, t=40, b=0),
        paper_bgcolor='#1e1e2e',
        font=dict(color='#cdd6f4', family=GLOBAL_FONT),
        legend=dict(
            bgcolor='#1e1e2e',
            bordercolor='#45475a',
            borderwidth=1,
            font=dict(color='#cdd6f4', family=GLOBAL_FONT),
            title=dict(text='Contributing Cause', font=dict(color='#cdd6f4', family=GLOBAL_FONT)),
        ),
    )
    # Style the animation slider and play button
    fig.update_layout(
        sliders=[{
            'bgcolor':     '#1e1e2e',
            'bordercolor': '#45475a',
            'font':        {'color': '#cdd6f4', 'size': 12},
            'currentvalue': {'font': {'color': '#cdd6f4', 'size': 14}},
            'tickcolor':   '#cdd6f4',
        }],
        updatemenus=[{
            'bgcolor':     '#313244',
            'bordercolor': '#45475a',
            'font':        {'color': '#cdd6f4'},
        }],
    )
    # ── Color slider step labels by season via HTML font tags ──────────
    WINTER = {'01 Jan', '02 Feb', '03 Mar', '11 Nov', '12 Dec'}
    WINTER_COLOR = '#7eb8f7'
    SUMMER_COLOR = '#f6c600'
    new_steps = []
    for slider in fig.layout.sliders:
        for step in slider.steps:
            color = WINTER_COLOR if step.label in WINTER else SUMMER_COLOR
            new_steps.append(dict(
                args=step.args,
                label=f'<span style="color:{color}">{step.label}</span>',
                method=step.method,
                value=step.value,
            ))
        slider.update(steps=new_steps)
    return fig


df_map['marker_size'] = (1 + (df_map['NUMBER OF PERSONS INJURED'] * 3).clip(upper=10))

# ── hover column names
df_map = df_map.rename(columns={
    'NUMBER OF PERSONS INJURED': 'Injured',
    'NUMBER OF PERSONS KILLED':  'Killed',
})
```

```python
# render the visualzation
fig_cause = px.scatter_mapbox(
    df_map,
    lat='latitude',
    lon='longitude',
    color='cause',
    color_discrete_map=CAUSE_COLORS,
    animation_frame='month_label',
    size='marker_size',
    size_max=11,
    hover_data={
        'latitude':    False,
        'longitude':   False,
        'month_label': False,
        'marker_size': False,
        'severity':    True,
        'BOROUGH':     True,
        'cause':       True,
        'Injured':     True,
        'Killed':      True,
    },
    zoom=MAP_ZOOM,
    # title
    title='NYC Vehicle Collisions by Contributing Cause (2014–2022)',
    subtitle='Dot color = cause       Dot size = persons injured        Slider:  winter (Nov–Mar)  summer (Apr–Oct)',
    labels={'cause': 'Contributing Cause', 'severity': 'Severity',
            'BOROUGH': 'Borough', 'Injured': 'Persons Injured', 'Killed': 'Persons Killed'},
)

fig_cause.update_traces(marker=dict(opacity=0.7))
fig_cause = apply_layout(fig_cause)
fig_cause.show()
```
